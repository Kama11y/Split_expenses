```python
expense_list = []
user_dict = {}
```

```python
def add_expense(name, amount, paid_by, *users, split="Equal", desc=""):
    global user_dict
    global expense_list
    users = list(users)
    expense_list.append([name,amount,paid_by,users,split,desc])
    if split.strip().lower() == "equal":
        split_amt = amount/len(users)
        for user in users: 
            if user.lower().strip() not in user_dict:
                user_dict[user.lower().strip()] = 0
            if user.lower().strip() == paid_by.lower().strip():
                user_dict[user.lower().strip()] = user_dict[user.lower().strip()] + amount - split_amt
            else:
                user_dict[user.lower().strip()] = user_dict[user.lower().strip()] - split_amt
    else:
        weights = list(map(float, input("Enter weights separated by space: ").split()))
        split_amt = [weights[i]*amount for i in range(len(weights))]
        for i in range(len(users)): 
            if users[i].lower().strip() not in user_dict:
                user_dict[users[i].lower().strip()] = 0
            if users[i].lower().strip() == paid_by.lower().strip():
                user_dict[users[i].lower().strip()] = user_dict[users[i].lower().strip()] + amount - split_amt[i]
            else:
                user_dict[users[i].lower().strip()] = user_dict[users[i].lower().strip()] - split_amt[i]
```

```python
def net_balances():
    global user_dict
    print("Balances :")
    for i in user_dict:
        if user_dict[i] > 0:
            print(i.capitalize(), "Should Receive", user_dict[i])
        else:
            print(i.capitalize(),"Owes", -1*user_dict[i])
```

```python
def settle_expenses():
    global user_dict
    creditors = []
    debitors = []
    for user in user_dict:
        if user_dict[user] > 0:
            creditors.append(user)
        elif user_dict[user] < 0:
            debitors.append(user)
        else:
            print(user.capitalize(), "Doesn't have to settle:)")
    i = 0
    j = 0

    while i < len(debitors) and j < len(creditors):
        debtor, debt_amt = debitors[i], user_dict[debitors[i]]
        creditor, credit_amt = creditors[j], user_dict[creditors[j]]

        settle_amt = min(debt_amt, credit_amt)

        print(f"{debtor.capitalize()} pays {creditor.capitalize()} ₹{-1*settle_amt:.2f}")

        user_dict[debitors[i]] -= settle_amt
        user_dict[creditors[j]] -= settle_amt

        if user_dict[debitors[i]] == 0:
            i += 1
        if user_dict[creditors[j]] == 0:
            j += 1
        
```

```python
add_expense("Hotel", 1200,"Alice","Alice","Bob","Carol")
```

```python
add_expense("Dinner", 600, "Bob","Alice","Bob","Carol")
```

```python
add_expense("Breakfast", 300, "Carol","Alice","Bob","Carol")
```

```python
net_balances()
```

```python
settle_expenses()
```
